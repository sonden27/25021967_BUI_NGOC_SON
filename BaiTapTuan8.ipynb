```python
#W7A16

def main(a):
    a.sort()
    cnt = 1; res = 0

    for i in range(1, len(a)):
        if a[i] != a[i-1]:
            if cnt < a[i-1]: res += cnt
            else: res += cnt - a[i-1]
            cnt = 1
        else: cnt += 1
    
    if cnt < a[len(a) - 1]: res += cnt
    else: res += cnt - a[len(a) - 1]

    if res == 0: print("YES")
    else: print("NO", res)

while 1:
    try:
        a = list(map(int, input().split()))
        main(a)
    
    except EOFError:
        break
```

```python
#W7A20

def count_digit(n):
    if n == 0: return 1
    cnt = 0
    while n > 0:
        cnt += 1
        n //= 10
    return cnt

def main(a, x):
    res = 0
    for i in a:
        if count_digit(i) == x: res += i
    print(res)


while 1:
    try:
        a = list(map(int, input().split()))
        x = int(input())
        main(a, x)
    
    except EOFError:
        break
```

```python
#W7A26

def add(cnt, x):
    if cnt == -1: cnt = int(x)
    else: cnt = cnt * 10 + int(x)
    return cnt

def main(a):
    numbers = []
    cnt = -1
    for i in range(len(a)):
        if (ord('0') <= ord(a[i]) <= ord('9')): cnt = add(cnt, a[i])
        else:
            if cnt != -1: numbers.append(cnt)
            cnt = -1
    if cnt != -1: numbers.append(cnt)

    total = 0
    for i in numbers: total += i
    print(total, total / len(numbers))



while 1:
    try:
        a = input()
        main(a)
    
    except EOFError:
        break
```

```python
#W7A27

open_bracket = ['(', '{', '[', '<']
close_bracket = [')', '}', ']', '>']

def main(a):
    st = []
    for i in a:
        if i in open_bracket: st.append(i); continue
        j = close_bracket.index(i)
        if (len(st) == 0 or st[len(st) - 1] != open_bracket[j]): return 0
        st.pop()
    
    if len(st) > 0: return 0
    else: return 1



while 1:
    try:
        a = input()
        if main(a) == 0: print('NO')
        else: print('YES')
    
    except EOFError:
        break
```

```python
#W7A28

def update(res, l, r):
    if res[0] < r - l + 1: res = (r - l + 1, l, r)
    return res

def main(n):
    check = [1] * (n + 1)
    check[0] = check[1] = 0
    for i in range(2, n + 1):
        if check[i] == 1:
            for j in range(i * i, n + 1, i):
                check[j] = 0
    
    j = 4; res = (-1, 0, -1)
    for i in range(2, n):
        if check[i] == 0: continue
        else:
            res = update(res, j, i - 1)
            j = i + 1
    if check[n - 1] == 0: res = update(res, j, n - 1)
    
    for i in range(res[1], res[2]): print(i, end = ', ')
    
    print(res[2])



while 1:
    try:
        n = int(input())
        main(n)
    
    except EOFError:
        break
```

```python
#W7A30

def bs(a, l, r, x):
    res = r + 1
    while l <= r:
        mid = (l + r) // 2
        if a[mid] >= x:
            res = mid
            r = mid - 1
        else: l = mid + 1
    return res

def main(a):
    b = [-1] * (len(a) + 1)
    res = 0
    for i in range(len(a)):
        pos = bs(b, 0, res, a[i])
        res = max(res, pos)
        b[pos] = a[i]
    
    for i in range(1, res + 1): print(b[i], end = ' ')
    print('')

        
while 1:
    try:
        a = list(map(int, input().split()))
        main(a)


    except EOFError:
        break
```
