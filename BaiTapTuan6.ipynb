```python
#W6A9
A=list(map(str,input().split()))
B=list(map(str,input().split()))
C={}
for i in A:
    a,b=i.split(':')
    if(a not in C):
        C[a]=int(b)
    else:
        C[a]+=int(b)
for i in B:
    a,b=i.split(':')
    if(a not in C):
        C[a]=int(b)
    else:
        C[a]+=int(b)
C=sorted(C.items())
print(C)
```

```python
#W6A10
A=list(map(int,input().split()))
k=int(input())
R=[]
for i in range(0,len(A)-1,1):
    for j in range(i+1,len(A),1):
        if(A[i]+A[j]==k):
            R.append((A[i],A[j]))
R.sort(key=lambda x: x[0])
print(R)
```

```python
#W6A12
A=list(map(int,input().split()))
B={}
for i in A:
    if(i not in B):
        B[i]=1
    else:
        B[i]+=1
B=dict(sorted(B.items(), key=lambda item: item[1], reverse=True))
min_val=0
max_count=0
flag=False
for i in B:
    if (not flag):
        min_val = i
        max_count = B[i]
        flag=True
    else:
        if(B[i] != max_count):
            break
        else:
            min_val = min(min_val, i)
print(min_val)
```
