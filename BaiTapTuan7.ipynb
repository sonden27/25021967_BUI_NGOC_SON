```python
#W7A1
def binary_search(arr, tar):
    low=0
    high=len(arr) - 1
    while(low<=high):
        mid= low + (high - low)//2
        if(arr[mid]==tar):
            return mid
        if (tar>arr[mid]):
            low = mid + 1
        else:
            high = mid -1
    return -1

a= list(map(int,input().split()))
x= int(input())
print(binary_search(a,x))
```

```python
#W7A8
def main(s):
    p = s.strip('[]').split('], [')
    a = []
    for z in p:
        tmp = tuple(map(int, z.split(',')))
        l = tmp[0]; r = tmp[1]
        a.append((r - l + 1, l, r))

    s = eval(input())
    b = []
    for i in range(len(s)):
        b.append((s[i], i))

    #Include all result by -1
    res = [-1] * len(b)
    

    #Sort intevals and queries
    a.sort()
    b.sort()

    #DSU
    par = [0] * (len(b) + 1)
    for i in range(len(b) + 1):
        par[i] = i

    def find_set(v):
        if v == par[v]: return v
        par[v] = find_set(par[v])
        return par[v]
    
    def union_sets(u, v):
        a = find_set(u)
        b = find_set(v)
        if a != b: par[a] = b

    #Binary search left
    def bsl(b, l, r, x):
        res = -1
        while l <= r:
            mid = (l + r) // 2
            if b[mid][0] >= x:
                res = mid
                r = mid - 1
            else: l = mid + 1
        return res
            

    #Binary search right
    def bsr(b, l, r, x):
        res = -1
        while l <= r:
            mid = (l + r) // 2
            if b[mid][0] <= x:
                res = mid
                l = mid + 1
            else: r = mid - 1
        return res
    
    for length, left, right in a:
        l = bsl(b, 0, len(b) - 1, left)
        r = bsr(b, 0, len(b) - 1, right)
        if r < l: continue

        #Brute force all query founded
        #DSU to skip all query updated answer before
        cur = find_set(l)
        while cur <= r:
            pos = b[cur][1]
            res[pos] = length
            union_sets(cur, cur + 1)
            cur = find_set(cur)
        
    print(res)


    
        
while 1:
    try:
        s = input()
        main(s)


    except EOFError:
        break
```

```python
#W7A9
def inp(s):
    s = s.strip('[]').split(', ')
    a = []
    for i in s:
        a.append(int(i))
    return a

def main(a):
    ma = 0
    for i in a: ma = max(ma, i)
    b = [[0 for _ in range(ma + 1)] for _ in range(len(a))]
    for i in range(len(a)):
        b[i][a[i]] += 1
    dp = b
    
    for j in range(1, ma + 1):
        dp[len(a)-1][j] += dp[len(a)-1][j-1]
    for i in range(len(a) - 2, -1, -1):
        for j in range(1, ma + 1):
            dp[i][j] += dp[i+1][j] + dp[i][j-1] - dp[i+1][j-1]

    res = [0] * len(a)
    for i in range(len(a)):
        res[i] = dp[i][a[i] - 1]
    
    print(res)

while 1:
    try:
        s = input()
        a = inp(s)
        main(a)


    except EOFError:
        break
```

```python
#W7A10
def inp(s):
    s = s.strip('[]').split(', ')
    a = []
    for i in s:
        a.append(int(i))
    return a

def bs(a, l, r, x):
    res = r + 1
    while l <= r:
        mid = (l + r) // 2
        if a[mid] > x:
            res = mid
            r = mid - 1
        else: l = mid + 1
    return res

def main(a, m):
    dp = [0] * len(a)
    dp[0] = a[0] % m
    for i in range(1, len(a)):
        dp[i] = (dp[i-1] + a[i] % m) % m
    
    #Using list same function as set
    res = dp[0]
    b = [dp[0]]
    for i in range(1, len(a)):
        res = max(res, dp[i])
        pos = bs(b, 0, i - 1, dp[i])
        if pos == i: b.append(dp[i])
        else:
            res = max(res, (dp[i] - b[pos] + m) % m)
            b.insert(pos, dp[i])
    
    print(res)

while 1:
    try:
        s = input()
        a = inp(s)
        x = int(input())
        main(a, x)


    except EOFError:
        break
```
