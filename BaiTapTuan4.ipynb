```python
#W4A13
from math import sqrt
a,b=map(int,input().split())
def tong_uoc(x):
    sum=1
    for i in range(2,int(sqrt(x)+1),1):
        flag=0
        s=i
        while(x%i==0):
            flag=1
            x//=i
            s*=i
        if(flag!=0):
            sum*=(s-1)//(i-1)
    if(x!=1):
        sum*=(x*x-1)//(x-1)
    return sum
if(a==tong_uoc(b)-b and b==tong_uoc(a)-a):
    print("true")
else:
    print("false")
```
