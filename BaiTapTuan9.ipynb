```python
#W9A4

def main(a, b):
    cnt_a = 0
    for i in range(len(a)):
        if ord('A') <= ord(a[i]) <= ord('Z'): cnt_a += 1

    j = 0
    cnt_b = 0
    pos = [] #tuple of b[i], index b[i], index a[j]
    for i in range(len(b)):
        while j < len(a) and ord('a') <= ord(a[j]) <= ord('z'):
            j += 1
        
        if j < len(a) and b[i] == a[j]:
            cnt_b += 1
            pos.append((b[i], i, j))
            j += 1
        elif i == 0 and len(pos) == 0:
            pos.append((-1, -1, -1))
    
    pos.append((-1, len(b), len(a)))
    
        
    if cnt_a != cnt_b: return 0

    a = a.upper()
    for i in range(1, len(pos)):
        left = pos[i-1][2] + 1
        right = pos[i][2] - 1
        for j in range(pos[i-1][1] + 1, pos[i][1]):
            check = 0

            for k in range(left, right + 1):
                if a[k] == b[j]:
                    check = 1
                    left = k + 1
                    break
            
            if check == 0: return 0
    
    return 1
            


while 1:
    try:
        t = int(input())
        while t > 0:
            a = input()
            b = input()
            if (main(a, b) == 0): print('NO')
            else: print('YES')
            t -= 1
        
        print('')

    
    except EOFError:
        break
```

```python
#W9A6

def next_permutation(a):
    i = len(a) - 2
    #Find the largest index i satisfy a[i] < a[i+1]
    while i >= 0 and a[i] >= a[i+1]: i -= 1

    #If no index was found, the permutation is last one (descending order)
    if i == -1:
        a.reverse()
        return a

    #Find the largest j greater than i satisfy a[j] > a[i]
    j = len(a) - 1
    while j > i and a[j] <= a[i]: j -= 1

    #Swap a[i], a[j]
    a[i], a[j] = a[j], a[i]

    #Reverse the sequence a[i+1] to final
    a[i+1:] = reversed(a[i+1:])

    return a

while 1:
    try:
        n = int(input())
        a = list(map(int, input().split()))
        res = next_permutation(a)
        for i in res: print(i, end = ' ')
        print('')
    
    except EOFError:
        break
```

```python
#W9A7

def canPlaceFlowers(a, k):
    if len(a) == 1 and a[0] == 0:
        k -= 1
    
    if len(a) == 1: return (k <= 0)

    for i in range(len(a)):
        if 0 < i < len(a) - 1 and a[i-1] == a[i] == a[i+1] == 0:
            a[i] = 1; k -= 1
        elif i == 0 and a[i] == a[i+1] == 0:
            a[i] = 1; k -= 1
        elif i == len(a) - 1 and a[i-1] == a[i] == 0:
            a[i] = 1; k -= 1
    
    return (k <= 0)

while 1:
    try:
        a = list(map(int, input().split()))
        k = int(input())
        print(canPlaceFlowers(a, k))
    
    except EOFError:
        break
```

```python
#W9A8
move = ['L', 'R', 'U', 'D']
move_x = [0, 0, 1, -1]
move_y = [-1, 1, 0, 0]

def is_path_crossing(moves):
    trace = [(0, 0)]
    curpos = (0, 0)
    for i in range(len(moves)):
        z = move.index(moves[i])
        curpos = (curpos[0] + move_x[z], curpos[1] + move_y[z])
        if curpos in trace: return True
        trace.append(curpos)
    
    return False


while 1:
    try:
        moves = input()
        print(is_path_crossing(moves))
    
    except EOFError:
        break
```

```python
#W9A9

def maxx(a, b):
    if a[0] > b[0]: return a
    else: return b

def minn(a, b):
    if a[0] < b[0]: return a
    else: return b

def main(a):
    ma1 = (int(-1e18), -1); mi1 = (int(1e18), -1)
    for i in range(len(a)):
        ma1 = maxx(ma1, (a[i], i))
        mi1 = minn(mi1, (a[i], i))
    
    ma2 = (int(-1e18), -1); mi2 = (int(1e18), -1)
    for i in range(len(a)):
        if (i != ma1[1]):
            ma2 = maxx(ma2, (a[i], i))
        if (i != mi1[1]):
            mi2 = minn(mi2, (a[i], i))
    
    for i in range(len(a)):
        if i != ma1[1] and i != ma2[1] and i != mi1[1] and i != mi2[1]:
            print(i)
            return

while 1:
    try:
        a = list(map(int, input().split()))
        main(a)
    
    except EOFError:
        break
```

```python
#W9A10

def check(a, b, r):
    pre = [0] * (len(a) + 1)
    left = 0; right = 0
    for i in range(len(b)):
        while right < len(a) and a[right] <= b[i] + r:
            right += 1
        while left < right and a[left] < b[i] - r:
            left += 1
        # print(left, right)
        pre[left] += 1; pre[right] -= 1
    
    for i in range(1, len(a)):
        pre[i] += pre[i-1]
    
    for i in range(len(a)):
        if pre[i] == 0: return 0
    
    return 1

def bs(a, b, l, r):
    res = r + 1
    while l <= r:
        mid = (l + r) // 2
        if check(a, b, mid) == 1:
            res = mid
            r = mid - 1
        else: l = mid + 1
    
    return res


def main(a, b):
    a.sort(); b.sort()
    print(bs(a, b, 0, int(1e9)))
    # print(check(a, b, 0))

while 1:
    try:
        a = list(map(int, input().split()))
        b = list(map(int, input().split()))
        main(a, b)
    
    except EOFError:
        break
```
